Q1. What is Flask Framework? What are the advantages of Flask Framework?

7. Large Community: Flask has a large community of developers who contribute to its development and create a wide range of plugins, libraries, and resources. This community support helps developers solve problems quickly and efficiently.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the mechanism by which a URL is mapped to a particular view function. A view function is a Python function that handles a specific URL request and returns a response. App routing in Flask is implemented using the @app.route() decorator.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run(debug=True)
```

In this code, we have defined two routes: / and /about. When a user visits the homepage (i.e., /), the home() function will be called and it will return the message "Hello, World!". Similarly, when a user visits the URL /about, the about() function will be called and it will return the message "This is the about page.".

App routing is an essential feature of Flask because it allows developers to create a URL structure for their web application. It provides a way to organize the application's functionality into discrete views or pages. App routing makes it easy for developers to add, modify, and remove pages from their application without having to modify the application's core logic.

Some of the reasons why we use app routes in Flask are:

1. Clear URL structure: App routing allows developers to create a clear and consistent URL structure for their web application. This helps users to understand and navigate the application easily.

2. Separation of concerns: App routing allows developers to separate the different parts of the application's functionality into discrete views or pages. This separation of concerns makes the application easier to maintain and modify.

3. Reusability: App routing allows developers to reuse the same view function for multiple URLs. This can be useful when implementing functionality that is common to multiple pages, such as a header or footer.

4. Modularity: App routing makes it easy to add, modify, and remove pages from the application without having to modify the application's core logic. This modularity helps to keep the codebase organized and maintainable.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. The url_for() function generates a URL for a specific function in the application. It takes the name of the function as the first argument and any additional arguments needed to generate the URL.

Here's an example code that demonstrates the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the homepage.'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/contact')
def contact():
    return 'This is the contact page.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
```

In this code, we defined three routes: /, /about, and /contact. We then used the url_for() function to generate URLs for these routes.

When the code is executed, the url_for() function generates the URLs for each route and prints them to the console. The test_request_context() function is used to create a test request context so that we can use the url_for() function outside of the normal request handling flow.

output is:

```python
/
/about
/contact
```

This code demonstrates how the url_for() function can be used to generate URLs for specific routes in Flask. You can use the generated URLs in your templates or anywhere else in your application where you need to link to a specific route.
